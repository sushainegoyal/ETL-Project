# Get news from news API

```python
last_month= pd.date_range('2019-11-17', periods=30, freq='D')
df = pd.DataFrame({ 'Date': last_month})
df.head()
```

```python
params = {
    "q": "Tesla",
    "sortBy": "relevancy",
    "language":"en",
    "pageSize":10,
    "apiKey": "ce1dfdc73ace4300a3c7ba5d0d1b8b96"}
```

```python
title = []
desc = []
```

```python
#for each day in last month pull news headline and desciption from the news api
for index, row in df.iterrows():
    base_url = 'https://newsapi.org/v2/everything?'
    date =row["Date"]
    next_date = date+dt.timedelta(days=1)
    print(date)
    # update address key value
    params["from"] = date
    params["to"] = next_date
    print(params)
    # make request
    result = requests.get(base_url, params=params).json()
    for i in range(10):
        title.append(result["articles"][i]["title"])
        desc.append(result["articles"][i]["description"])
```

```python
# back up list to avoid overwrite
print(len(title))
print(len(desc))
title_backup=title
desc_backup=desc
```

```python
def chunks(lst, n):
    """Yield successive n-sized chunks from lst."""
    for i in range(0, len(lst), n):
        yield lst[i:i + n]
        
title_chunked = list(chunks(title, 10))
print(title_chunked[0])
desc_chunked = list(chunks(desc, 10))
```
